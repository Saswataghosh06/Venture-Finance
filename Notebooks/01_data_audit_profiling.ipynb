```python
import pandas as pd
import numpy as np

# 1. Load the "Bronze" (Raw) Data
# Note: Using 'r' before the string to handle backslashes safely
df_cust = pd.read_csv(r"D:\FinTech\Data\Raw\customers.csv")
df_tx = pd.read_csv(r"D:\FinTech\Data\Raw\transactions.csv")

def perform_comprehensive_audit(df, table_name):
    # FIX: Swapped outer quotes to single ' so inner " works for the math
    print(f"{'='*20} AUDIT REPORT: {table_name} {'='*20}")
    
    # 2. Basic Dimensions
    print(f"Total Records: {df.shape[0]}")
    print(f"Total Columns: {df.shape[1]}")
    
    # 3. Data Type Check
    print("\n[1] Data Types & Memory:")
    print(df.dtypes)
    
    # 4. Null Value Analysis
    null_summary = df.isnull().sum()
    null_perc = (df.isnull().sum() / len(df)) * 100
    # Creating a cleaner report table
    null_report = pd.DataFrame({'Missing Values': null_summary, 'Percentage (%)': null_perc})
    print("\n[2] Null Value Analysis:")
    print(null_report[null_report['Missing Values'] > 0])
    
    # 5. Duplicate Detection
    total_dupes = df.duplicated().sum()
    print(f"\n[3] Row-Level Duplicates: {total_dupes}")
    
    # 6. ID Uniqueness (Crucial for FinTech)
    # Identifying which ID column to check based on the dataframe
    id_col = 'cust_id' if 'cust_id' in df.columns else 'tx_id'
    
    if id_col in df.columns:
        unique_ids = df[id_col].nunique()
        print(f"[4] Unique IDs ({id_col}): {unique_ids}")
        if unique_ids < (len(df) - total_dupes):
            print("    WARNING: You have overlapping IDs with different data (Logic Error)!")
    else:
        print(f"[4] Unique ID Check: Primary ID column not found.")

    # 7. Categorical Messiness (String Profiling)
    if 'category' in df.columns:
        print("\n[5] Categorical Variation (Category):")
        print(df['category'].value_counts())
        
    # 8. Numerical Outlier Detection
    if 'amount' in df.columns:
        print("\n[6] Financial Stats (Outlier Check):")
        stats = df['amount'].agg(['min', 'max', 'mean', 'median'])
        print(stats)

# --- Run the Audit ---
# Make sure these variables match exactly what you defined in Step 1
perform_comprehensive_audit(df_cust, "Customers")
print("\n" + "-"*60 + "\n")
perform_comprehensive_audit(df_tx, "Transactions")
```
